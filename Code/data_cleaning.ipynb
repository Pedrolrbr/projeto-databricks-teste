```python
import random
from datetime import datetime, timedelta
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType

# Gerar dados aleatórios
num_rows = 100

# Listas para dados categóricos
categorias = ['A', 'B', 'C', 'D']
cidades = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Brasília', 'Salvador']
status = ['Ativo', 'Inativo', 'Pendente']

# Gerar dados
data = []
for i in range(num_rows):
    data.append((
        i + 1,  # id
        f"Usuario_{random.randint(1000, 9999)}",  # nome
        random.choice(categorias),  # categoria
        random.choice(cidades),  # cidade
        random.randint(18, 70),  # idade
        round(random.uniform(1000, 10000), 2),  # valor
        random.choice(status),  # status
        (datetime(2023, 1, 1) + timedelta(days=random.randint(0, 365))).date()  # data
    ))

# Definir schema
schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("nome", StringType(), True),
    StructField("categoria", StringType(), True),
    StructField("cidade", StringType(), True),
    StructField("idade", IntegerType(), True),
    StructField("valor", DoubleType(), True),
    StructField("status", StringType(), True),
    StructField("data", DateType(), True)
])

# Criar DataFrame
df_test = spark.createDataFrame(data, schema)

print(f"DataFrame criado com {df_test.count()} registros")
display(df_test)
```
